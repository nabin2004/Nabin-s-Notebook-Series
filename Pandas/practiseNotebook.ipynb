```python
def pct30(column):
    return column.quantile(0.3)

def pct40(column):
    return column.quantile(0.4)
```

```python
df['avg_price'].agg([pct30,pct40])
```

```python
df['avg_price'].cumsum()
```

```python
df['avg_price'].cumprod()
```
